###`Q1:` Join Tuples if similar initial element
While working with Python tuples, we can have a problem in which we need to perform concatenation of records from the similarity of initial element. This problem can have applications in data domains such as Data Science.

For eg.
```
Input  : test_list = [(5, 6), (5, 7), (5, 8), (6, 10), (7, 13)]
Output : [(5, 6, 7, 8), (6, 10), (7, 13)]
```

````python
# write your code here
test_list = [(5, 6), (5, 7), (5, 8), (6, 10), (7, 13)]

res = {}

for i, j in test_list:
    if i in res:
        # If the key already exists, append the new element to the list
        res[i].append(j)
    else:
        # If the key is new, create a list starting with both the key and the first element
        res[i] = [i, j]

output = list(res.values())
print(res)
print(output)

output = [tuple(x) for x in output]

print(output)
````

###`Q2:` Multiply Adjacent elements (both side) and take sum of right and lest side multiplication result.


For eg.
```
The original tuple : (1, 5, 7, 8, 10)
Resultant tuple after multiplication :

(1*5, 1*5+5*7, 7*5 + 7*8, 8*7 + 8*10, 10*8) -> (5, 40, 91, 136, 80)

output-(5, 40, 91, 136, 80)
```

````python
# write your code here
t1 = (1,5,7,8,10)
result = []
for i in range(0,len(t1)+1):
  if i == 0:
    result.append(t1[i] * t1[i+1])
  elif i == len(t1)-1:
    result.append(t1[i] * t1[i-1])
    break
  else:
    result.append(t1[i] * t1[i-1] + t1[i] * t1[i+1])

result = tuple(result)
print(result)

````

###`Q4`: Count no of tuples, list and set from a list
```
list1 = [{'hi', 'bye'},{'Geeks', 'forGeeks'},('a', 'b'),['hi', 'bye'],['a', 'b']]

```
`Output:`

```
List-2
Set-2
Tuples-1
```

````python
# write your code here

list1 = [{'hi', 'bye'}, {'Geeks', 'forGeeks'}, ('a', 'b'), ['hi', 'bye'], ['a', 'b']]

list_count = 0
set_count = 0
tuple_count = 0

for item in list1:

    if type(item) == list:
        list_count += 1

    elif type(item) == set:
        set_count += 1

    elif type(item) == tuple:
        tuple_count += 1

print("List-", list_count)
print("Set-", set_count)
print("Tuples-", tuple_count)
````

```

        ar2 = [6, 7, 20, 80, 100]
```

### `Q3`: Convert List to List of dictionaries. Given list values and keys list, convert these values to key value pairs in form of list of dictionaries.

**Example 1:**

Input:
```bash
test_list = ["DataScience", 3, "is", 8]
key_list = ["name", "id"]
```

Output:

```bash
[{'name': 'DataScience', 'id': 3}, {'name': 'is', 'id': 8}]
```

**Example 2:**

Input:
```bash
test_list = ["CampusX", 10]
key_list = ["name", "id"]
```

Output:

```bash
[{'name': 'CampusX', 'id': 10}]
```

````python
# write your code here
test_list = ["CampusX", 10]
key_list = ["name", "id"]

result = []

for i in range(0, len(test_list), 2):
    temp_dict = {}

    temp_dict[key_list[0]] = test_list[i]
    temp_dict[key_list[1]] = test_list[i+1]

    result.append(temp_dict)

print(result)
````

### `Q5`: Sort Dictionary key and values List.


Output:

```bash

```

**Example 2:**

Input:

```bash
{'c': [10, 34, 3]}
```

Output:

```bash
{'c': [3, 10, 34]}
```

````python
# write your code here
data = {'c': [3], 'b': [12, 10], 'a': [19, 4]}

result = {}

for key in sorted(data.keys()):   # step 1: sort keys
    result[key] = sorted(data[key])   # step 2: sort values

print(result)

````
