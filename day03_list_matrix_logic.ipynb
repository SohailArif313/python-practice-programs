###`Problem 1:` Combine two lists index-wise(columns wise)

Write a program to add two lists index-wise. Create a new list that contains the 0th index item from both the list, then the 1st index item, and so on till the last element. any leftover items will get added at the end of the new list.

`Given List:`
```
list1 = ["M", "na", "i", "Kh"]
list2 = ["y", "me", "s", "an"]
```

`Output:`
```
[['M','y'], ['na', me'], ['i', 's'], ['Kh', 'an']]
```

````python
# write code here
list1 = ["M", "na", "i", "Kh"]
list2 = ["y", "me", "s", "an"]

result = []

for a, b in zip(list1, list2):
    result.append([a, b])

print(result)
````

###`Problem 5:` You are given a list of integers. You are asked to make a list by running through elements of the list by adding all elements greater and itself.

i.e. Say given list is `[2,4,6,10,1]`
resultant list will be `[22,20,10,23]`.

For 1st element `2` ->> these are greater `(4+6+10)` values and `2` itself so on adding becomes `22`.

For 2nd element `4` ->> greater elements are `(6, 10)` and `4` itself, so on adding `20`

like wise for all other elememts.

`[2,4,6,10,1]`-->`[22,20,16,10,23]`




````python
# Write code here
lst = [2, 4, 6, 10, 1]
result = []

for i in lst:
    total = i  # start with itself
    for j in lst:
        if j > i:
            total += j
    result.append(total)

print(result)
````

###`Problem 6:` Find list of common unique items from two list. and show in increasing order

`Input`

```
num1 = [23,45,67,78,89,34]
num2 = [34,89,55,56,39,67]
```

`Output:`
```
[34, 67, 89]
```


````python
num1 = [23,45,67,78,89,34]
num2 = [34,89,55,56,39,67]
common = []

for i in num1:
    if i in num2 and i not in common:
        common.append(i)

common.sort()
print(common)
````

###`Problem 7:` Sort a list of alphanumeric strings based on product value of numeric character in it. If in any string there is no numeric character take it's product value as 1.

`Input:`
```
['1ac21', '23fg', '456', '098d','1','kls']
```

`Output:`
```
['456', '23fg', '1ac21', '1', 'kls', '098d']

```

````python
# Write code here
lst = ['1ac21', '23fg', '456', '098d','1','kls']
prod = 1
prod_list = []
for i in lst:
  for j in i:
    if j.isdigit():
      prod *= int(j)
  prod_list.append(prod)
  prod = 1

result = [x for _,x in sorted(zip(prod_list,lst),reverse=True)]
print(result)


````

### `Problem 10:` Add Space between Potential Words.

**Example:**

Input:

```bash
['campusxIs', 'bestFor', 'dataScientist']
```

Output:
```bash
['campusx Is', 'best For', 'data Scientist']
```

````python
# write code here
words = ['campusxIs', 'bestFor', 'dataScientist']

result = []

for word in words:
    new_word = ""

    for ch in word:
        if ch.isupper():
            new_word += " "   # add space before uppercase
        new_word += ch

    result.append(new_word)

print(result)
````

### `Problem 11:` Write a program that can perform union operation on 2 lists

**Example:**

Input:

```bash
[1,2,3,4,5,1]
[2,3,5,7,8]
```

Output:
```bash
[1,2,3,4,5,7,8]
```

````python
# Write code here
list1 = [1,2,3,4,5,1]
list2 = [2,3,5,7,8]

result = []

# add from list1
for num in list1:
    if num not in result:
        result.append(num)

# add from list2
for num in list2:
    if num not in result:
        result.append(num)

print(result)
````

### `Problem 14:` Write a list comprehension that can transpose a given matrix

matrix = [<br>
  [1,2,3],<br>
          [4,5,6],<br>
          [7,8,9]<br>]<br>

[1, 4, 7]<br>
[2, 5, 8]<br>
[3, 6, 9]<br>

````python
# write code here
matrix = [
  [1, 2, 3],
  [4, 5, 6],
  [7, 8, 9]
]

# Transpose using list comprehension
transpose = [[row[i] for row in matrix] for i in range(3)]
for row in transpose:
    print(row)
````
